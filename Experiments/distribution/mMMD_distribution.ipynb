```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import multivariate_normal
from utils import *

# Define sample sizes and number of runs
n_sample_size = 200
num_runs = 2000
figure_size = (3.5,2.5)
bins = 30
```

```python
from scipy.stats import multivariate_normal

dim1 = 20
dim2 = 200

x_mean1 = np.zeros(dim1); y_mean1 = np.zeros(dim1)
x_cov1 = y_cov1 = np.eye(dim1)

X_distribution1 = multivariate_normal(x_mean1, x_cov1)
Y_distribution1 = multivariate_normal(y_mean1, y_cov1)
empirical_stats1 = empirical_mMMD_distribution(X_distribution1, Y_distribution1, kernel = RBF_Kernel, n_sample_size = n_sample_size, num_runs = num_runs, n_jobs=8, random_state = 42)

x_mean2 = np.zeros(dim2); y_mean2 = np.zeros(dim2)
x_cov2 = y_cov2 = np.eye(dim2)

X_distribution2 = multivariate_normal(x_mean2, x_cov2)
Y_distribution2 = multivariate_normal(y_mean2, y_cov2)
empirical_stats2 = empirical_mMMD_distribution(X_distribution2, Y_distribution2, kernel = RBF_Kernel, n_sample_size = n_sample_size, num_runs = num_runs, n_jobs=8, random_state = 42)

plot_overlayed_distributions(empirical_stats1, empirical_stats2, figure_size = figure_size, bins = 25, label1 = "d = 20", label2 = "d = 200", title = "Empirical Distributions of mMMD", save_path = "results/mMMD_normal_d_20_200_k_gauss_hist.pdf")
```
